```python
def crawlHouse(url):
    
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--no-sandbox')
    capabilities = chrome_options.to_capabilities()
    
    browser = webdriver.Chrome(r"C:\Users\PC\Documents\Python Scripts\BienesRaices\chromedriver.exe", desired_capabilities=capabilities) #, chrome_options=chrome_options)
    browser.get(url)
    
    titulo = browser.find_element_by_css_selector('h1.item-title__primary').text.replace(',',' ')
    tcaracteristicas = browser.find_elements_by_xpath('//section/ul/li[@class="specs-item"]/strong')
    vcaracteristicas = browser.find_elements_by_xpath('//section/ul/li[@class="specs-item"]/span')

    caracteristicas = [x.text for x in tcaracteristicas]
    valores = [x.text for x in vcaracteristicas]

    metroscuadradostotales = float(valores[caracteristicas.index('Superficie total')].replace('m²', '')) if 'Superficie total' in caracteristicas else 1
    cuartos = int(valores[caracteristicas.index('Cuartos')]) if 'Cuartos' in caracteristicas else 0
    cuartosbano = int(valores[caracteristicas.index('Baños')]) if 'Baños' in caracteristicas else 0
    antiguedad = valores[caracteristicas.index('Antigüedad')] if 'Antigüedad' in caracteristicas else 0

    precio = float(browser.find_element_by_css_selector('span.price-tag-fraction').text.replace('.',''))

    try:
        ubicacion = browser.find_elements_by_xpath('//div[2]/img')
        lot = ubicacion[1].get_attribute('srcset')
        lot = lot[lot.find("center=")+7:lot.find("&zoom")].replace('%2C',',').split(',')
        lat,lon = float(lot[0]),float(lot[1])
    except:
        print("no location")
        nolocation.append(url)
        browser.close()
        return
    
    print('titulo: ',titulo)
    print('metroscuadradostotales: ', metroscuadradostotales)
    print('cuartos: ',cuartos)
    print('cuartosbano: ', cuartosbano)
    print('antiguedad: ', antiguedad)
    print('latitud: ',lat)
    print('longitud: ',lon)
    print('precio: ', precio)
    
    lsttowrite = [titulo,antiguedad,metroscuadradostotales,cuartos,cuartosbano,'-',precio,'-','-',lat,lon,precio/metroscuadradostotales,'-']
    linetowrite = [str(x) for x in lsttowrite]

    with open('casas.csv','a') as file: 
        file.write('|'.join(linetowrite)+'\n')

#     except:
#         print('Something else happened')
#         nofetch.append(url)
#     finally:
        
    browser.close()
```

```python
def crawlLinks(url, classa='item__info-link'):
    
    response = ''
    while response == '':
        try:
            response = requests.get(url)
        except:
            print("Connection refused by the server..")
            print("Let me sleep for 5 seconds")
            print("ZZzzzz...")
            time.sleep(5)
            print("Was a nice sleep, now let me continue...")
            continue
    
    soup = BeautifulSoup(response.text, "html.parser")
    items = soup.select('a.'+classa)
    for item in items:
        print(item["href"])
        crawlHouse(item["href"])
```

```python
base = 'https://inmuebles.mercadolibre.com.ec/casas/venta/casas-guayaquil'

skip = 1
extra = ''

while True:
    url = base+extra
    print(url)
    
    if skip < 200:
        crawlLinks(url)
    else:
        break
    
    skip += 50   
    extra = '_Desde_'+str(skip)
    
```
